## 6. Data Visualisation

#### importing data visualization package seaborn


```python
import seaborn as sns
```

#### ploting graph between age and count of people having diabetes

```python
sns.countplot(data=df,x="Age",hue="Diabetes",palette="rocket")
```

#### ploting graph between gender and count of people having diabetes

```python
sns.countplot(data=df,x="Sex",hue="Diabetes",palette="mako")
```

#### ploting graph between age,sex and count of people having diabetes

```python
grouped = df.groupby(['Age', 'Sex', 'Diabetes']).size().reset_index(name='Count')

# Plotting using Seaborn
plt.figure(figsize=(12, 6))

# Create the bar plot
sns.barplot(data=grouped, x='Age', y='Count', hue='Sex', palette='pastel', ci=None)

# Adding title and labels
plt.title('Count of Individuals with Diabetes by Age and Gender')
plt.xlabel('Age')
plt.ylabel('Count')

# Show plot
plt.legend(title='Sex')
plt.show()
```

#### ploting graph between BMI and count of people having diabetes

```python
plt.figure(figsize=(25, 16))
sns.countplot(data=df,x="BMI",hue="Diabetes",palette="mako",width=1)
plt.xticks(rotation=90)
plt.tight_layout()
```
